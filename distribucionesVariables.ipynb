```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min, skewnorm
```

# Edades

```python
# Definir los parámetros de la distribución
mu = 55  # media
sigma = 10  # desviación estándar
a = 1  # asimetría

# Generar 1000 datos aleatorios que siguen la distribución
data = skewnorm.rvs(a, loc=mu, scale=sigma, size=1000)
```

```python
plt.hist(data, bins=15)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma de datos generados")
plt.xticks(np.arange(18, 71, 5))
plt.show()
```
